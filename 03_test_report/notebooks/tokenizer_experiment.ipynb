# 🧪 토크나이저(Tokenizer) 비교 실험

이 노트북은 Kiwi 형태소 분석기를 사용하여 **기본 사전(Default)**과 **사용자 정의 도메인 사전(Custom Domain Dictionary, V3)** 간의 토큰화 성능 차이를 비교하는 실험 공간입니다.

## 설정
이 노트북이 `skn-third-proj` 가상환경에서 실행되고 있는지 확인하세요.

```python
import os
import sys
from kiwipiepy import Kiwi

# 1. 프로젝트 루트 경로 설정 로직
# 노트북이 src/notebooks/ 하위에 있다고 가정하고 루트를 찾습니다.
current_path = os.getcwd()
project_root = current_path

# 'data' 폴더가 보일 때까지 상위 디렉토리로 이동
while not os.path.exists(os.path.join(project_root, 'data')) and project_root != '/':
    project_root = os.path.dirname(project_root)

if project_root == '/':
    # Fallback: src/notebooks 에 있다고 가정하고 두 단계 상위로
    project_root = os.path.abspath(os.path.join(os.getcwd(), "../../"))
    
sys.path.append(project_root)
print(f"📂 Project Root: {project_root}")

DICT_PATH = os.path.join(project_root, "src/core/tokenizer/domain_dictionary.txt")
if os.path.exists(DICT_PATH):
    print(f"✅ 도메인 사전을 찾았습니다: {DICT_PATH}")
else:
    print(f"❌ 도메인 사전을 찾을 수 없습니다: {DICT_PATH}")
```

## 토크나이저 초기화
두 가지 Kiwi 인스턴스를 생성합니다:
1. `kiwi_default`: 순정 Kiwi (기본 모델)
2. `kiwi_custom`: `domain_dictionary.txt`가 로드된 커스텀 Kiwi

```python
# 기본 Kiwi 초기화
kiwi_default = Kiwi()

# 커스텀 Kiwi 초기화
kiwi_custom = Kiwi()
try:
    kiwi_custom.load_user_dictionary(DICT_PATH)
    print("✨ 커스텀 사용자 정의 사전 로드 완료")
except Exception as e:
    print(f"⚠️ 사전 로드 실패: {e}")
```

## 비교 함수
아래 셀을 실행하여 비교용 함수를 정의하세요.

```python
def compare_tokens(text):
    tokens_def = [t.form for t in kiwi_default.tokenize(text)]
    tokens_cus = [t.form for t in kiwi_custom.tokenize(text)]
    
    print(f"📝 입력 문장: {text}")
    print(f"   ❌ Default: {tokens_def}")
    print(f"   ✅ Custom : {tokens_cus}")
    
    # 결과가 다를 경우 강조 표시
    if tokens_def != tokens_cus:
        print("   💡 차이가 발견되었습니다!")
    else:
        print("   (결과가 동일합니다)")
    print("-" * 50)
```

## 실험 공간
아래 리스트를 수정하여 직접 문장을 테스트해보세요!

```python
# 테스트 케이스
examples = [
    "집사는 맛동산을 캐기 위해 벤토나이트 모래를 샀다.",
    "메인쿤과 아비시니안은 성격이 다르다.",
    "고양이 췌장염은 재발이 잦다.",
    "츄르를 먹는 먼치킨 고양이",
    "러시안블루는 조용한 편이다."
]

for ex in examples:
    compare_tokens(ex)
```
